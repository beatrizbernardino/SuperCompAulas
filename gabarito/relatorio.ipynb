```python
import subprocess
import time
import random
import numpy as np
import matplotlib.pyplot as plt
```

Algortitmo Guloso: Heuristica mais leve

```python
W=330

for i in range(100,3000,200):
    file = 'in-{0}.txt'.format(i)  # nome do arquivo a ser gerado
    f = open(file, 'w')
    f.writelines(f'{i} {W} \n')
    for j in range(i):
        peso= np.random.randint(1,31)
        valor=np.random.randint(1,11)
        f.writelines(f'{peso} {valor} \n')
    f.close()
    print(''.join(file))
```

```python
mais_leve=[]
n=[]
for i in range(100,3000,200):

    with open('in-{0}.txt'.format(i)) as f:
        start = time.perf_counter()
        proc = subprocess.run(['./mochila_mais_leve'], input=f.read(), text=True, capture_output=True)
        end = time.perf_counter()

        # print('Saída:', proc.stdout)
        # print('Stderr:', proc.stderr)
        print('Tempo total(s):', end - start)
        mais_leve.append(end - start)
        n.append(i)
```
